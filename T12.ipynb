```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import  rpy2
```

```python
import rpy2.robjects as ro

import rpy2.robjects.numpy2ri
```

```python
rpy2.robjects.numpy2ri.activate()
```

```python
codigo_r = """
saludar <-function(cadena){
return(paste("Hola,",cadena))
}
"""
```

```python
ro.r(codigo_r)
```

```python
saludar_py = ro.globalenv["saludar"]
```

```python
res = saludar_py("Hola antonio Banderas")
res[0]
```

```python
var_from_python = ro.FloatVector(np.arange(1,5,0.1))
```

```python
var_from_python
```

```python
print(var_from_python.r_repr())
```

```python
ro.globalenv["var_to_r"] = var_from_python
```

```python
ro.r("var_to_r")
```

```python
np.sum(var_from_python)
```

```python
ro.r("summary(var_to_r)")
```
