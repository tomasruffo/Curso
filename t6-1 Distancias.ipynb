# Clustering jerarquico 



```python
import  matplotlib.pyplot as ply
from scipy.cluster.hierarchy import dendrogram , linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward")
```

```python
Z
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendograma jerarquico para el clustering")
plt.ylabel("ID de los usuarios de Netflix")
plt.xlabel("Distancia")
dendrogram(Z , leaf_rotation=90, leaf_font_size=0.8)
plt.show()
           
```

```python
Z = linkage(data[movies], method="single", metric="cosine")
plt.figure(figsize=(25,10))
plt.title("Dendograma jerarquico para el clustering")
plt.ylabel("ID de los usuarios de Netflix")
plt.xlabel("Distancia")
dendrogram(Z , leaf_rotation=90, leaf_font_size=0.8)
plt.show()
```
