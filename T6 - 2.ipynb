# Clustering jerarquico y dendrogramas

Antes de empezar, pongamos un poco de rotacion para hablar todos el mismo idioma

* X data set(array de n x m) de puntos a clusterizar
* n numeros de datos
* m numero de rasgos
* Z array de enlace de cluster con la informacion de las uniones
* k numero de clusters

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram ,linkage
import numpy as np
```

```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0],[[3,1], [1,4]], size = [100,])
b = np.random.multivariate_normal([0,20], [[3,1],[1,4]], size = [50,])
X = np.concatenate((a,b))

print (X.shape)
plt.scatter(X[:,0], X[:,1])
plt.show()
                  
```

```python
Z = linkage(X, "ward")
Z
```

### Dendograma tuneado 

```python
def dendogram_tune(*args, **kwargs):
    import pandas as pd
    max_d=kwargs.pop("max_d", None)
    if max_d and "color_threshold" not in kwargs:
        kwargs["color_threshold"] = max_d
    annotate_above = kwargs.pop("annotate_above", 0)
    
    ddata= dendogram(*args, **kwargs)
    
    if not kwargs.get("no_plot", False):
        plt.title('Clustering jerarquico ')
        plt.xlabel("indice del data set ( o del cluster)")
        plt.ylabel("Distancia")
           for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
                x = 0.5 * sum(i[1:3])
                y= d[1]
                if y > annotate_above:
                    plt.plot(x,y,"o", c =c )
                    plt.annotate('%.3g'%y, (x,y), xytext=(0,-5),

                                textcoords="offset points", va='top', ha='center')
        
            if max_d:
                plt.axhline(y==max_d, c='k')
                
            return ddata    
```
